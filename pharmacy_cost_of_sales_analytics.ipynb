##Aggregation function by month of sums of costs of sales

```python
def aggregate_by_month_sums(cost_of_sales_df):
  # Assuming you have your dataframe named 'df'
  # First, convert the 'DATE' column to a datetime format if it's not already
  cost_of_sales_df['DATE'] = pd.to_datetime(cost_of_sales_df['DATE'], format='%d/%m/%Y')

  # Extract the year and month from the 'DATE' column
  cost_of_sales_df['YEAR'] = cost_of_sales_df['DATE'].dt.year
  cost_of_sales_df['MONTH'] = cost_of_sales_df['DATE'].dt.month

  # Group by 'YEAR', 'MONTH' and sum the 'EXPENSE' column
  aggregated_df = cost_of_sales_df.groupby(['YEAR', 'MONTH'])['COST'].sum().reset_index()

  # Display the aggregated dataframe
  return aggregated_df
```

```python
cost_of_sales_monthly_sums_df = aggregate_by_month_sums(cost_of_sales_df)
```

##Function that calculates monthly sums for specific SALE INVOICE CATEGORY and YEAR

```python
def filter_by_year_payer_monthly_sums(df, year, sale_invoice_category):
    # Filter the dataframe by year and social security payer
    filtered_df = df[(df['DATE'].dt.year == year) & (df['SALE INVOICE CATEGORY'] == sale_invoice_category)]

    # Group the filtered dataframe by month and calculate the sum of 'INVOICE' column
    grouped_df = filtered_df.groupby(df['DATE'].dt.month)['COST'].sum().reset_index()

    # Rename columns for clarity
    grouped_df.columns = ['Month', 'Monthly Sum']
    return grouped_df
```

```python
year = 2022
sale_invoice_category = 'ΙΔΙΩΤΕΣ ΠΡΟΜΗΘΕΥΤΕΣ'
result_df = filter_by_year_payer_monthly_sums(cost_of_sales_df, year, sale_invoice_category)
print(result_df)
```

#Function that calculates dataframe of specific monthly sales category

```python
def filter_by_monthly_sums(df, sale_invoice_category):
    # Filter the dataframe by year and social security payer
    filtered_df = (df['SALE INVOICE CATEGORY'] == sale_invoice_category)

    # Group the filtered dataframe by month and calculate the sum of 'INVOICE' column
    grouped_df = filtered_df.groupby(['COST'].sum().reset_index()

    # Rename columns for clarity
    grouped_df.columns = ['Month', 'Monthly Sum']
    return grouped_df
```

```python
sale_invoice_category = 'ΙΔΙΩΤΕΣ ΠΡΟΜΗΘΕΥΤΕΣ'
result_df = filter_by_year_payer_monthly_sums(cost_of_sales_df, year, sale_invoice_category)
print(result_df)
```
