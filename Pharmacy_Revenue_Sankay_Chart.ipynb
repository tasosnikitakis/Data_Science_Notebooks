```python
import pandas as pd
import plotly.graph_objects as go
import kaleido
import urllib, json
```

```python
revenue_inflow_df = pd.read_csv("/content/drive/MyDrive/pharmacy_data/revenue_stream.csv")
```

```python
revenue_outflow_df = pd.read_csv("/content/drive/MyDrive/pharmacy_data/revenue_outflow.csv")
```

```python
df = pd.read_csv("/content/drive/MyDrive/pharmacy_data/df.csv")
```

```python
df
```

```python
revenue_inflow_df
```

```python
revenue_outflow_df
```

```python
links = pd.concat([revenue_inflow_df, revenue_outflow_df], axis=0)
links
```

```python
unique_source_target = list(pd.unique(links[['source','target']].values.ravel('k')))
```

```python
unique_source_target
```

```python
mapping_dict = {k: v for v, k in enumerate(unique_source_target)}
```

```python
mapping_dict
```

```python
links['source'] = links['source'].map(mapping_dict)
links['target'] = links['target'].map(mapping_dict)
```

```python
links
```

```python
links_dict = links.to_dict(orient='list')
```

```python
links_dict
```

```python
fig = go.Figure(data=[go.Sankey(
    node = dict(
        pad = 15,
        thickness=20,
        line=dict(color='black', width=0.5),
        label = unique_source_target,
        color='green'
    ),
    link = dict(
        source= links_dict['source'],
        target = links_dict['target'],
        value = links_dict['value']
    )

)
])
fig.update_layout(title='Revenue Sankey Chart')
fig.show()
```

```python
unique_source_target = list(pd.unique(df[['source','target']].values.ravel('k')))
```
