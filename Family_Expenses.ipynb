#Cosmote Eurobank credit card expenses

```python
eurobank_df = pd.read_csv("statement.csv", sep=";")
```

```python
eurobank_df.columns
```

```python
eurobank_df.drop(['ΔΟΣΗ', 'ΕΞΟΔΑ  ΣΥΝΑΛΛΑΓΜΑΤΟΣ  (€)', 'ΕΞΟΔΑ  ΑΝΑΛΗΨΗΣ  ΜΕΤΡΗΤΩΝ'], axis='columns', inplace=True)
```

```python
eurobank_df = eurobank_df[eurobank_df['ΤΥΠΟΣ'] == 'ΑΓΟΡΑ']
```

```python
eurobank_df['ΗΜ/ΝΙΑ  ΕΓΓΡΑΦΗΣ'] = pd.to_datetime(eurobank_df['ΗΜ/ΝΙΑ  ΕΓΓΡΑΦΗΣ'], format="%d/%m/%Y")
```

```python
eurobank_df
```

```python
eurobank_df['ΠΟΣΟ'] = eurobank_df['ΠΟΣΟ'].str.replace(',', '.')
```

```python
eurobank_df['ΠΟΣΟ'] = eurobank_df['ΠΟΣΟ'].astype(float)
```

```python
eurobank_card_sum = round(eurobank_df['ΠΟΣΟ'].sum(), 2)
```

```python
eurobank_card_sum
```
