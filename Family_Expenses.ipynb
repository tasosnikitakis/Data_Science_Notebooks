```python
import pandas as pd
```

#Toshl data

```python
# data loading into dataframe
toshl_df = pd.read_csv("Toshl_data.csv")
```

```python
toshl_df.columns
```

```python
# data cleaning and rearangements
toshl_df_cleared = toshl_df[["Date", "Tags", "Expense amount" ]]
toshl_df_cleared.rename(columns = {'Tags':'Category'}, inplace = True)
# transforming date column to datetime object values
toshl_df_cleared["Date"] = pd.to_datetime(toshl_df_cleared["Date"], format="%d/%m/%Y")
```

```python
toshl_df_cleared
```

```python
# Calculating sum of expenses
toshl_sum = toshl_df_cleared['Expense amount'].sum()
```

```python
toshl_sum
```

#Cosmote Eurobank credit card expenses

```python
# data loading into dataframe
eurobank_df = pd.read_csv("statement.csv", sep=";")
```

```python
eurobank_df.columns
```

```python
# data cleaning and rearangements
eurobank_df.drop(['ΗΜ/ΝΙΑ ΣΥΝΑΛΛΑΓΗΣ', 'ΔΟΣΗ', 'ΕΞΟΔΑ  ΣΥΝΑΛΛΑΓΜΑΤΟΣ  (€)', 'ΕΞΟΔΑ  ΑΝΑΛΗΨΗΣ  ΜΕΤΡΗΤΩΝ'], axis='columns', inplace=True)
# droping credit card payments
eurobank_df = eurobank_df[eurobank_df['ΤΥΠΟΣ'] == 'ΑΓΟΡΑ']
eurobank_df.drop(['ΤΥΠΟΣ'], axis='columns', inplace=True)
# transforming date column to datetime object values
eurobank_df['ΗΜ/ΝΙΑ  ΕΓΓΡΑΦΗΣ'] = pd.to_datetime(eurobank_df['ΗΜ/ΝΙΑ  ΕΓΓΡΑΦΗΣ'], format="%d/%m/%Y")
# transforming expenses column to float values instead of strings
eurobank_df['ΠΟΣΟ'] = eurobank_df['ΠΟΣΟ'].str.replace(',', '.')
eurobank_df['ΠΟΣΟ'] = eurobank_df['ΠΟΣΟ'].astype(float)
# Calculating sum of expenses
eurobank_card_sum = round(eurobank_df['ΠΟΣΟ'].sum(), 2)
```

```python
eurobank_df
```

```python
eurobank_card_sum
```

#Winbank credit card expenses

```python
# data loading into dataframe
wb_df = pd.read_csv("winbank_card.csv")
```

```python
# data cleaning and rearangements
wb_df.drop(['Ημ/νία Συναλλαγής', 'Αριθμός Παραστατικού', 'Νόμισμα'], axis='columns', inplace=True)
wb_df.drop(wb_df[wb_df['Κατηγορία'] == 'Ανακατανομή'].index, inplace = True)
# Calculating sum of expenses
winbank_credit_card_sum = round(wb_df['Ποσό'].sum(), 2)
```

```python
wb_df
```

```python
winbank_credit_card_sum
```

#NBG Bank credit card expenses

```python
# data loading into dataframe
nbg_df = pd.read_excel("CardStatementExport.xlsx")
```

```python
nbg_df.columns
```

```python
# data cleaning and rearangements
nbg_df.drop(['Αριθμός Κάρτας', 'Περίοδος Από', 'Περίοδος Έως', 'Α/Α', 'Αριθμός Κάρτας', 'Νόμισμα Λογαριασμού', 'Ποσό', 'Νόμισμα Συναλλαγής', 'Ισοτιμία Συναλλαγής', 'Αριθμός Κάρτας.1', 'Στοιχεία Εμπόρου'], axis='columns', inplace=True)
# data rearangements so as to turn date column into datetime object values
nbg_df['Ημερομηνία/Ώρα Συναλλαγής'].astype(str)
nbg_df[["date", "time", "am-pm"]] = nbg_df['Ημερομηνία/Ώρα Συναλλαγής'].str.split(pat=' ', expand=True)
nbg_df.drop(['Ημερομηνία/Ώρα Συναλλαγής', 'time', 'am-pm'], axis='columns', inplace=True)
# column rearangements
nbg_df = nbg_df.reindex(columns=['date', 'Περιγραφή Κίνησης', 'Χ/Π', 'Ποσό εντολής'])
# droping credit card payments
nbg_df = nbg_df[nbg_df['Χ/Π'] == 'Χ']
nbg_df.drop(['Χ/Π'], axis='columns', inplace=True)
# Calculating sum of expenses
nbg_credit_card_sum = round(nbg_df['Ποσό εντολής'].sum(), 2)
```

```python
nbg_df
```

```python
nbg_credit_card_sum
```
