```python
import pandas as pd
import random
```

```python
# List of player names
player_names = ["Player 1", "Player 2", "Player 3", "Player 4", "Player 5",
                "Player 6", "Player 7", "Player 8", "Player 9", "Player 10",
                "Player 11", "Player 12", "Player 13", "Player 14", "Player 15",
                "Player 16", "Player 17", "Player 18", "Player 19", "Player 20",
                "Player 21", "Player 22", "Player 23", "Player 24", "Player 25",
                "Player 26", "Player 27"]

# List of preferred positions
positions = ["Attacker", "Midfielder", "Defender", "Goalkeeper"]

# Generate random match appearances between 0 and 20 for each player
match_appearances = [random.randint(0, 20) for _ in range(27)]

# Create a dataframe
data = {'Name': player_names, 'Preferred Position': [random.choice(positions) for _ in range(27)], 'Match Appearances': match_appearances}
df = pd.DataFrame(data)
```
